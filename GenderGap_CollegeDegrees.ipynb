```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(9, 6))

for sp in range(0,6):
    ax = fig.add_subplot(2,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(16,16))

#Printing charts in the first row for stem categories
for x in range(0,18,3):
    ind=int(x/3)
    ax = fig.add_subplot(6,3,x+1)
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[ind]], c=cb_dark_blue, label="Women", linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[stem_cats[ind]], c=cb_orange, label="Men", linewidth=3)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=1)
    if ind != 5:
        ax.tick_params(labelbottom = 'off')
    ax.tick_params(top="off", bottom="off", left="off", right="off")
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1970,2015)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[ind])
    if ind == 0:
        ax.text(2000,80,"Women")
        ax.text(2000,10,"Men")
    elif ind == 5:
        ax.text(2000,10,"Women")
        ax.text(2000,85,"Men")
        
#Printing charts in the third row for other categories
for y in range(0,18,3):
    ind=int(y/3)
    ax = fig.add_subplot(6,3,y+3)
    ax.plot(women_degrees["Year"], women_degrees[other_cats[ind]], c=cb_dark_blue, label="Women", linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[other_cats[ind]], c=cb_orange, label="Men", linewidth=3)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.5)
    if ind != 5:
        ax.tick_params(labelbottom = 'off')
    ax.tick_params(top="off", bottom="off", left="off", right="off")
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1970,2015)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[ind])
    if ind == 0:
        ax.text(2000,80,"Women")
        ax.text(2000,20,"Men")
    elif ind == 5:
        ax.text(2000,10,"Women")
        ax.text(2000,85,"Men")
        
#Printing charts in the second row for liberal arts categories
for z in range(0,15,3):
    ind=int(z/3)
    ax = fig.add_subplot(6,3,z+2)
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[ind]], c=cb_dark_blue, label="Women", linewidth=3)
    ax.plot(women_degrees["Year"], 100-women_degrees[lib_arts_cats[ind]], c=cb_orange, label="Men", linewidth=3)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.3)
    if ind != 4:
        ax.tick_params(labelbottom = 'off')
    ax.tick_params(top="off", bottom="off", left="off", right="off")
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1970,2015)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[ind])
    if ind == 0:
        ax.text(1970,80,"Women")
        ax.text(1970,10,"Men")
    elif ind == 4:
        ax.text(1970,20,"Women")
        ax.text(1970,70,"Men")
plt.show()
```
